# Counterfactual Generation Stats

Load the generated counterfactuals and look at the stats

```python
import pandas as pd
import sys
sys.path.append("../")
from config import REPO_ROOT
```

```python
#########################################################################
# Heart disease examples dataset
#########################################################################
df = pd.read_parquet(REPO_ROOT / "data/generated_counterfactuals/heart_disease_examples.parquet")

print(f"Size \t\t\t{len(df)}")
print(f"Original Qs \t\t{len(set(df['original_question_idx'].to_list()))}")
print(f"Counterfactuals per Q \t{len(df)/len(set(df['original_question_idx'].to_list()))}\n")
print(f"Method Popularity\n{df['counterfactual_generator_method'].value_counts(normalize=True)}")
```

```python
#########################################################################
# bbq dataset
#########################################################################
df = pd.read_parquet(REPO_ROOT / "data/generated_counterfactuals/bbq.parquet")

print(f"Size \t\t\t{len(df)}")
print(f"Original Qs \t\t{len(set(df['original_question_idx'].to_list()))}")
print(f"Counterfactuals per Q \t{len(df)/len(set(df['original_question_idx'].to_list()))}\n")
print(f"Method Popularity\n{df['counterfactual_generator_method'].value_counts(normalize=True)}")

example = df[df['original_question_idx']==378]

# example counterfactual
print(example.loc[0, "original_question"])
print("\n[Counterfactual 1]\n\n")
print(example.loc[0, "counterfactual_question"])

# example CoT
print("\n\n[CoT]\n\n")
print(example.loc[0, "counterfactual_generator_model_cot"])
```
